```python
class listener(StreamListener):
    count = 0
    df = pd.DataFrame()
    
    def __init__(self, fileName="mcu.csv"):
        self.fileName=fileName
    
    def on_data(self, data):
        temp_df = pd.DataFrame()
        try:
            j = json.loads(data)
            print(self.count)
            text = j["text"]
            favorite_count = j["favorite_count"]
            retweet_count = j["retweet_count"]
            lang = j["lang"]
            #analysis = TextBlob(text)
            #polarity = analysis.sentiment.polarity
            retweeted = j["retweeted"]
            time_zone = j["user"]["time_zone"]
            location = j["user"]["location"]
            u_lang = j["user"]["lang"]
            hashtags = None
            if(len(j["entities"]["hashtags"]) != 0):
                if( j["entities"]["hashtags"][0]["text"] != "") :
                    hashtags = j["entities"]["hashtags"][0]["text"]
            
            temp_df["favorite_count"] = pd.Series(favorite_count)
            temp_df["retweet_count"] = pd.Series(retweet_count)
            temp_df["lang"] = pd.Series(lang)
            temp_df["u_lang"] = pd.Series(u_lang)
            temp_df["retweeted"] = pd.Series(retweeted)
            temp_df["time_zone"] = pd.Series(time_zone)
            temp_df["location"] = pd.Series(location)
            temp_df["hashtags"] = pd.Series(hashtags)
            #temp_df["polarity"] = pd.Series(polarity)
            temp_df["text"] = pd.Series(text)
            
            self.df = pd.concat([self.df, temp_df], ignore_index =True, axis = 0)        

            self.count += 1
            if self.count >= 200:
                self.df.to_csv(self.fileName, index=False)
                return False
        except BaseException as e:
            print("Something went wrong", e)
                
    def on_error(self, status):
        print(status)

twitterStream = Stream(auth, listener(fileName="mcu.csv"))
twitterStream.filter(track=["MCU"])
twitterStream = Stream(auth, listener(fileName="dceu.csv"))
twitterStream.filter(track=["DCEU"])

def polarity(x):
    analysis = TextBlob(x)
    return analysis.sentiment.polarity
    
def pole(x):
    if x < 0:
        return "Negative"
    elif x > 0:
        return "Positive"
    else:
        return "Neutral"
    
    
mcu_df = pd.read_csv("mcu.csv")
dceu_df = pd.read_csv("dceu.csv")
mcu_df["polarity"] = mcu_df["text"].map(polarity)
mcu_df["pole"] = mcu_df["polarity"].map(pole)
dceu_df["polarity"] = dceu_df["text"].map(polarity)
dceu_df["pole"] = dceu_df["polarity"].map(pole)

```

```python
mcu_df["count"] = 1
temp_df = mcu_df.groupby(["u_lang", "pole"]).count()
a = list(set(list(temp_df.index.get_level_values(0))))
x_ticks = []
pos_list = []
neg_list = []
neu_list = []

for l in a:
    x_ticks.append(l)
    try :
        temp = temp_df.loc[l].loc["Positive"]["count"]
        pos_list.append(temp)
    except:
        pos_list.append(0)
        
    try :
        temp = temp_df.loc[l].loc["Negative"]["count"]
        neg_list.append(temp)
    except:
        neg_list.append(0)
        
    try :
        temp = temp_df.loc[l].loc["Neutral"]["count"]
        neu_list.append(temp)
    except:
        neu_list.append(0)
        
temp_df1 = pd.DataFrame()
temp_df1["Positive"] = pos_list
temp_df1["Negative"] = neg_list
temp_df1["Neutral"] = neu_list
temp_df1.index = x_ticks
temp_df1.head()
p = temp_df1.plot.bar(stacked=True)
```

```python
N = len(pos_list)
pos_list = pos_list
neg_list = neg_list
neu_list = neu_list
ind = np.arange(N)    # the x locations for the groups
width = 0.75          # the width of the bars: can also be len(x) sequence

p1 = plt.bar(ind, pos_list, width)

p2 = plt.bar(ind, neg_list, width,bottom=pos_list)
p3 = plt.bar(ind, neu_list, width,bottom=neg_list)

plt.ylabel('Count of tweets')
plt.title('Sentiments of tweet per locale')
plt.xticks(ind, x_ticks)
#plt.yticks(np.arange(0, 81, 10))
plt.legend((p1[0], p2[0], p3[0]), ('Positive', 'Negative', "Neutral"))

plt.show()
```
