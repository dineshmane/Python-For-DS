```python
import pandas as pd
from IPython.display import display, HTML
import numpy as np
#from TAS_Python_Utilities import data_viz
from sklearn import neural_network
from sklearn import metrics
import subprocess
from IPython.display import display, HTML, Image
import io as io
import matplotlib.pyplot as plt
from sklearn.tree import export_graphviz
from sklearn import preprocessing
from sklearn import linear_model
from sklearn import svm
from sklearn.model_selection import cross_val_score
```

### Loading dataset

```python
df = pd.read_csv("train.csv")
df_test = pd.read_csv("test.csv")
```

#### Pre-processing at Row-Level:
Here, we wanted to check how many rows have mostly NAs as their entries. 

Replace string NA values with pre-defined na values in python

```python
train = df
test = df_test
train_wNA=train.replace('na',np.nan)
test_wNA=test.replace('na',np.nan)
#train_wNA
```

```python
train_wNA.count()
train_wNA=train_wNA.dropna(how='all')
train_wNA

test.count()
test_wNA=test.dropna(how='all')
test_wNA.shape
```

Counting Number of NA values for attributes in each row

```python
train_cnt=train_wNA.isnull().sum(axis=1)
train_cnt=train_cnt.to_frame()

type(train_cnt)
list(train_cnt)
#abcd.columns = ['attr','count']

train_wNA.index.name = 'index_train'
train_cnt.index.name = 'train_cnt_index'


test_cnt=test_wNA.isnull().sum(axis=1)
test_cnt=test_cnt.to_frame()

type(train_cnt)
list(train_cnt)
#abcd.columns = ['attr','count']

test_wNA.index.name = 'index_train'
test_cnt.index.name = 'train_cnt_index'
```

Resultant final dataframe to be merged to train dataset

```python
final_res_train = pd.DataFrame()
final_res_train["count"]= train_cnt.iloc[:][0]

final_res_test = pd.DataFrame()
final_res_test["count"]= test_cnt.iloc[:][0]
```

Merging the frequency column to training and testing dataset for further filtering

```python
final_train=pd.concat([train_wNA, final_res_train], axis=1)
final_train = final_train[(final_train['count']<80)]
final_train = final_train[final_train.columns.difference(["count"])]
final_train.to_csv('row_filter_train.csv')

final_test=pd.concat([train_wNA, final_res_test], axis=1)
final_test = final_test[(final_test['count']<80)]
final_test = final_test[final_test.columns.difference(["count"])]
final_test.to_csv('row_filter_test.csv')
```

```python
df = final_train
df_test = final_test
```

### End of row level pre processing

```python
df.head()
```

Null check

```python
display(df[df.isnull().any(axis=1)])  
```

```python
df = df.replace(np.nan, 'na')
df_test = df_test.replace(np.nan, 'na')

a = df.select_dtypes(include=[np.object]).describe()
na_cols = a.columns[a.loc["top"][:] == 'na']
print("Number of columns having na values", len(na_cols))
```

### na_df which has only NA columns

```python
na_df = df[na_cols]
na_df.head()
```

#### Vizulazing the data distribution in dataframe
segregating the "na" and "non-na" values

```python
# Function used to display the data distribution
def data_viz(df):
    # Plot histograms for numeric columns
    for idx, col in enumerate(df.select_dtypes(include=[np.number]).keys()):
        df[col].plot.hist(title = col)
        plt.show()

    # Plot bar plots  for categorical/boolean columns        
    for idx, col in enumerate(df.select_dtypes(include=[np.object, np.bool]).keys()):
        # Only draw bar plots for variables with less than 30 levels
        if len(df[col].unique()) < 30:
            df[col].value_counts().plot.bar(title = col)
            plt.show()
```

```python
na_df = na_df[na_df.iloc[:][:] == "na"]
na_df1 = na_df
na_df1 = na_df1.fillna("non-na")
data_viz(na_df1)
```

```python
threshold_list =[]
for col in na_cols:
    if na_df1[col].value_counts()[1] > 15000:
        threshold_list.append(col)
print("These are the column names which has more NA values than threshold :", threshold_list)
```

Removing threshold columns from the dataset

```python
df2 = df[df.columns.difference(threshold_list)]
df2.head()
```

```python
X_train = df2[df2.columns.difference(["class"])]
Y_train = np.array(df2["class"])

# we have selected the columns which we have choosen for the train dataset
X_test = df_test[df2.columns.difference(["class"])]
Y_test = np.array(df_test["class"])
```

### Normalizing the input features

```python
X_train = X_train.replace("na",np.nan)
X_test = X_test.replace("na",np.nan)

X_train = X_train.astype('float64')
X_train = X_train.fillna(X_train.mean())

X_test = X_test.astype('float64')
X_test = X_test.fillna(X_test.mean())
```

```python
# Getting normalized data for train
x = X_train.values #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
X_train = pd.DataFrame(x_scaled)


x = X_test.values #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
X_test = pd.DataFrame(x_scaled)
```

### Evaluating Model Performance

#### Logistic Regression

```python
acc_dict = {}
```

```python
logistic = linear_model.LogisticRegression(C=1e5)
log_model = logistic.fit(X_train,Y_train)
```

```python
# Make a set of predictions for the test data
y_test = Y_test
y_pred = logistic.predict(X_test)

# Print performance details
accuracy = metrics.accuracy_score(y_test, y_pred) # , normalize=True, sample_weight=None
print("Accuracy: " +  str(accuracy))
acc_dict["logistic"] = accuracy
print(metrics.classification_report(y_test, y_pred))

# Print confusion matrix
print(metrics.confusion_matrix(y_test, y_pred))

# Print nicer confusion matrix
print("Confusion Matrix")
pd.crosstab(y_test, y_pred, rownames=['True'], colnames=['Predicted'], margins=True)
```

#### Neural Network

```python
my_nn = neural_network.MLPClassifier(activation='relu', alpha=0.0001, batch_size='auto', beta_1=0.9,
       beta_2=0.999, early_stopping=False, epsilon=1e-08,
       hidden_layer_sizes=(50,100,50,), momentum=0.9,
       shuffle=True, solver='adam', validation_fraction=0.1, warm_start=False)
my_nn.fit(X_train,Y_train)
```

```python
# Make a set of predictions for the test data
y_test = Y_test
y_pred = my_nn.predict(X_test)

# Print performance details
accuracy = metrics.accuracy_score(y_test, y_pred) # , normalize=True, sample_weight=None
print("Accuracy: " +  str(accuracy))
acc_dict["Neural Network"] = accuracy
print(metrics.classification_report(y_test, y_pred))

# Print confusion matrix
print(metrics.confusion_matrix(y_test, y_pred))

# Print nicer confusion matrix
print("Confusion Matrix")
pd.crosstab(y_test, y_pred, rownames=['True'], colnames=['Predicted'], margins=True)
```

### Support Vector Machine

### Graphs
