# Sentiment Analysis on Skype reviews from trustpilot.com website

### Imports

```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import urllib
import json
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import math
from wordcloud import WordCloud, STOPWORDS
from nltk import ngrams
from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')
import re
```

```python
basePage = 'http://www.trustpilot.com/review/'
reviewSite = 'www.skype.com'
mainLink = basePage + reviewSite
```

```python
r = urllib.request.urlopen(mainLink).read()
soup = BeautifulSoup(r, "html.parser")  # lxml
review_count = soup.find("h2", class_="header--inline").text
```

```python
num_of_reviews = int(review_count.strip().split(" ")[0].replace(",",""))
review_per_page = 20
num_of_pages = int(math.ceil(num_of_reviews/review_per_page))
outputList = []
```

```python
for page in range(1, num_of_pages+1):
    page_url = mainLink + "?page=" + str(page)
    r = urllib.request.urlopen(page_url).read()
    soup = BeautifulSoup(r, "html.parser")  # lxml
    
    # review content size will be max 20
    review_content = soup.find_all("section", class_="review__content") 
    
    for review in review_content:
        
        title = review.find("h2", class_="review-content__title").text.strip()
        body = review.find("p", class_="review-content__text").text.strip()
        date_json = review.find("div", class_="review-content-header__dates").text
        date_raw = json.loads(date_json)["publishedDate"].split("T")[0]
        rating_raw_class = review.find("div", class_="star-rating")["class"]
        rating = rating_raw_class[1].split("-")[-1]
        outputList.append([title, body, date_raw, rating])
```

```python
skype_df = pd.DataFrame(outputList, columns=["title", "body", "date", "rating"])
```

```python
skype_df = pd.read_csv("skypeData.csv", sep="\t")
```

### Stopwords

```python
cloud_stopwords = list(set(STOPWORDS))
nltk_stopwords = list(set(stopwords.words('english')))
all_stopwords = cloud_stopwords + nltk_stopwords
all_stopwords.append("skype")
all_stopwords.append("microsoft")
all_stopwords.append("a")
all_stopwords.append("i")
```

### Processing corpus

```python
def processText(review):
    reviewWordList = re.split("[ .,;:!?‘’``''@#$%^_&*()<>{}~\n\t\\\-]", review.lower())
    reviewWordList = [word for word in reviewWordList if word != ""]
    review = " ".join(reviewWordList)
    wordList = review.split()
    wordList = [word for word in wordList if word not in all_stopwords]
    return " ".join(wordList)

skype_df["body"] = skype_df["body"].apply(processText)
```

```python
pos = skype_df[skype_df["rating"] >= 4]
neg = skype_df[skype_df["rating"] <= 3]
```

### Positive

```python
pos_text = " ".join(b for b in pos["body"] ) # joining all the sentences in one corpus
```

```python
total_words = len(pos_text.split())
vocab_size = len(set(pos_text.split()))

print("total words in positve reviews are ", total_words)
print("unique words in positive reviews are ", vocab_size)
```

##### 1-Gram Positve wordcloud

```python
wordcloud = WordCloud(scale=7,stopwords=all_stopwords, background_color="white").generate(pos_text)
fig = plt.figure(1, figsize = (15, 15))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```

### Two gram

```python
def n_gram_list(sentence, n=2):
    n = n
    twoGram = ngrams(sentence.split(), n)
    final_op = []
    for grams in twoGram:
        gramList = list(grams)
        two_gram = "_".join(gramList)
        final_op.append(two_gram)
    return final_op
```

##### 2-Gram Positive wordcloud

```python
final_op = n_gram_list(pos_text, 2)
df=pd.DataFrame(final_op)
print("top 5 ngram\n", df[0].value_counts()[:5])
ngram_str = " ".join(final_op)

wordcloud = WordCloud(scale=3, stopwords=all_stopwords, background_color="white").generate(ngram_str)
fig = plt.figure(1, figsize = (15, 15))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```

##### 2-Gram Negative wordcloud

```python
final_op = n_gram_list(neg_text, 2)
df=pd.DataFrame(final_op)
print("top 5 ngram\n", df[0].value_counts()[:5])
ngram_str = " ".join(final_op)

wordcloud = WordCloud(scale=3, stopwords=all_stopwords, background_color="white").generate(ngram_str)
fig = plt.figure(1, figsize = (15, 15))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
