#### Sorting is must step for using the IndexSlice

```python
idx = pd.IndexSlice
mi = mi.sort_index()
mi.index.is_lexsorted()
```

```python
mi.head()
```

```python
mi.loc[idx[:, 1997:2004],:].head(10)
```

```python
# 1997:2004:2   -> 1997, 1999, 2001, 2003
mi.loc[idx["aaltoan01":"adamsgr02", 1997:2004:2], :]
```

```python
# when we want to select 1997 to 2004 with one year gap and then 2004 to 2008 with 2 year gap 
# we can make different get_locs() and combine the result and pass it to iloc
temp = mi.index.get_locs((
            idx["aaltoan01":"adamsgr02", 1997:2000]
        ))
print(temp)
temp1 = mi.index.get_locs((
    idx["aaltoan01":"adamsgr02", 2000:2004:2]
    ))
temp1
```

```python
# get_locs() gives us index positions so we used iloc instead of loc
mi.iloc[np.concatenate((temp, temp1)),:]
```

```python
def get_many_locs(df, slices):
    arr=np.empty(0, dtype="int");
    for s in slices:
        locs = df.index.get_locs((s))
        arr = np.concatenate((arr,locs))
    return arr
```

```python
locs = get_many_locs(mi,
             (
                 idx["aaltoan01":"adamsgr02", 1997:2000],
                 idx["aaltoan01":"adamsgr02", 2000:2004:2]
             )
)
locs
```

```python
mi.iloc[locs,:]
```

## Multi index on long and wide dataframe

```python
team_splits = team_splits.drop(columns="tmID")
team_splits = team_splits.set_index(["year", "name"])
```

```python
team_splits.head()
```

```python
months = team_splits.columns.map(lambda x: x[:3])
metrics = team_splits.columns.map(lambda x: x[3:])
print(months)
print(metrics)
```

```python
midx = pd.MultiIndex.from_arrays([months, metrics])
midx
```

```python
team_splits.columns = midx
team_splits.head()
```

```python
team_splits = team_splits.stack(level=0)
team_splits.head(10)
```

```python
team_splits = team_splits.swaplevel(1,2)
team_splits.head()
```

```python
team_splits = team_splits.reorder_levels([2,0,1])
team_splits.head()
```

```python
team_splits.index.levels[2].name = "month"
team_splits.head()
```

```python
team_splits.groupby(level=["name", "year"]).sum().head(10)
```

Here if we want to change long to wide dataframe then pivot will be bit challenging. But we can use the **unstack method**

```python
team_splits = team_splits.unstack(level=["year", "month"])
team_splits.head()
```

```python
team_splits.columns = team_splits.columns.reorder_levels([1,2,0])
team_splits.head()
```

```python
team_splits.columns = team_splits.columns.reorder_levels([2,0,1])
team_splits.head()
```
